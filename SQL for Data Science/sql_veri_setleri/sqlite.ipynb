## LIKE ILE SECIM ISLEMLERI

<ul>
    <li>'a%':"a" ile baslayan degerler</li>
    <li>'%a':"a" ile biten degerler</li>
    <li>'%or%':Herhangi bir yerinde 'or' olan degerler</li>
    <li>'_r%':Ikinci sirasinda "r" olan degerler</li>
    <li>'a%%':'a' ile baslayip en az 3 karakterden olusan degerler</li>
    <li>'a%o':'a' ile baslayip "o" ile biten degerler</li>
</ul>

```python
%%sql
Select * from Customers Limit 10
```

```python
%%sql
Select * from Customers Limit 10
```
